```python
import sklearn
import pandas as pd
import numpy
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("moviesv2.0s.csv")
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=["id", "url", "director", "contentRating", "description", "duration", "keywords"])
```

```python
df = df.dropna()
```

```python
df['genre'] = df['genre'].str.strip()
```

```python
def splitDataFrameList(df,target_column,separator):
    ''' df = dataframe to split,
    target_column = the column containing the values to split
    separator = the symbol used to perform the split
    returns: a dataframe with each entry for the target column separated, with each element moved into a new row. 
    The values in the other columns are duplicated across the newly divided rows.
    '''
    def splitListToRows(row,row_accumulator,target_column,separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)
    new_rows = []
    df.apply(splitListToRows,axis=1,args = (new_rows,target_column,separator))
    new_df = pd.DataFrame(new_rows)
    return new_df
```

```python
new_df = splitDataFrameList(df, "genre", ",")
```

```python
new_df['genre'] = new_df['genre'].str.strip()
```

```python
new_df['genre']
```

```python
genre_list = [s for s in new_df['genre'] if len(s) > 1]
new_df = new_df[new_df['genre'].isin(genre_list)]
new_df
```

```python
new_df.to_csv("movies_clean2.0.csv")
```
